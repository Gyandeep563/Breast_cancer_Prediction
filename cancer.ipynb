```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
X=data.data
y=data.target
```

```python
print(X.shape)
```

```python
import pandas as pd

df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

df.head()
```

```python
print(data.target_names)

# malignant means cancerous 
# benign means non can
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('target',axis=1)
y=df['target']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.4,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
lg.fit(X_train,y_train)
```

```python
y_pred=lg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
dy_pred=dt.predict(X_test)
```

```python
accuracy_score(dy_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rt=RandomForestClassifier()
```

```python
rt.fit(X_train,y_train)
```

```python
ry_pred=rt.predict(X_test)
```

```python
accuracy_score(ry_pred,y_test)
```

```python
importance_df = pd.DataFrame({
    'Feature': data.feature_names,
    'Importance': rt.feature_importances_
})

importance_df.sort_values(
    by='Importance',
    ascending=False
)
```

### With Pipelines With LogisticRegression

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
pipe=Pipeline([
    ('scaler',StandardScaler()),
    ('model',LogisticRegression()),
])
```

```python
pipe.fit(X_train,y_train)
```

```python
py_pred=pipe.predict(X_test)
```

```python
accuracy_score(py_pred,y_test)
```

### Pipeline with RandomForest


```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe=Pipeline([
    ('scaler',StandardScaler()),
    ('model',RandomForestClassifier())
])
```

```python
pipe.fit(X_train,y_train)
```

```python
py_pred=pipe.predict(X_test)
```

```python
accuracy_score(py_pred,y_test)
```

### GridSearch CV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
params={
    'n_estimators':[10,20,40,60],
    'max_depth':[3,5,8,None]
}
```

```python
grid=GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid=params,
    cv=6
)
```

```python
grid.fit(X_train,y_train)
```

```python
accuracy_score(grid.predict(X_test),y_test)
```

```python
print('Best_parameters',grid.best_params_)
print("best_score",grid.best_score_)
```

```python
print(grid.cv_results_['mean_test_score'])
```
